# 1. Import necessary dependencies

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
```

# 2. Import the required dataset

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

# 3. Visualize different columns

### A. Drawing a simple Boxplot with default parameters

```python
# Make default box plot
sns.violinplot(y=df["sepal_length"])
plt.show()
```

### B. Improving the graph with different parameters

* Specify the data and the x-variable.
* Set the color of the violin.
* Set the caps for the whiskers.
* Set the Whisker length.
* Show the outliers
* Add a title and axis labels.

```python
sns.violinplot(
    y=df["sepal_width"],
    color="red",
    inner="quartiles",  # Show quartiles within the violin
    bw=0.8,           # Adjust the bandwidth (smoothness)
    cut=0,            # Control whisker extension
)

plt.title("Violin Plot of Sepal Width")
plt.ylabel("Sepal Width (cm)")
plt.xlabel("")
plt.show()
```

### C. Visualize multiple Violin plot together

1. Define the columns to plot ( update based on your scenario )
2. Define the colors of each density plot ( update based on your choice/preference )
3. we created a 2 * 2 matrix with each figure as (14 , 10 ) , you can update each of them based on preference


```python
# Define the columns to plot
columns_to_plot = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Define colors for each violin plot
colors = ['red', 'green', 'orange', 'coral']  # Add more colors if needed

# Create a 2x2 subplot grid with a fixed figure size
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Iterate through the columns and plot each one in a separate subplot
for i, col in enumerate(columns_to_plot):
    # Determine the row and column for the current subplot
    row = i // 2
    col_num = i % 2

    # Select the appropriate subplot axis
    ax = axes[row, col_num]

    # Create the violin plot, using a different color for each
    sns.violinplot(data=df, y=df[col], color=colors[i], inner="quartiles", bw=0.8, cut=0, ax=ax)

    # Set the title and labels for the subplot
    ax.set_title(f"{col.title()} Distribution")
    ax.set_ylabel(f"{col.title()} (cm)")  # Changed from xlabel to ylabel
    ax.set_xlabel("")

# Adjust the layout to prevent overlapping titles/labels
plt.tight_layout()

# Display the plots
plt.show()
```

# 4. How to Interpret Violin plot :

Key Features to Look For

Shape:

* A violin plot displays the probability density of the data at different values. It's a combination of a box plot and a kernel density estimate.
* Symmetric: If the violin is symmetric around the median, the data is likely to be symmetrically distributed.

Skewed:

* Right-skewed (positive skew): The violin will be stretched more on the right side.
* Left-skewed (negative skew): The violin will be stretched more on the left side.
* Unimodal/Bimodal/Multimodal: The width of the violin shows peaks in the data. A violin can have multiple peaks, indicating multiple modes in the data.

Center:

* The median is shown by a line (or dot, depending on parameters) inside the violin.

Spread:

* The width of the violin plot indicates the spread or variability of the data. Wider sections of the violin represent higher probability densities, and thus a greater concentration of data points.
* The interquartile range (IQR) is also typically shown within the violin (often as a box).

Outliers:

* Violin plots can show outliers, though not as explicitly as box plots. The shape of the violin can reveal the presence of extreme values.

Gaps:

* Violin plots, like density plots, don't show gaps in the data in the same way histograms do. However, narrow sections of the violin can indicate areas of low data density.
