# 1. Import necessary dependencies

```python
# libraries & dataset
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
```

```python
# Load the titanic dataset
df = sns.load_dataset('titanic')
```

# 2. Create data visualizations using Scatter plot

### A. Create default scatter plot with "age" and "fare"

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))  # Adjust the figure size for better readability
sns.scatterplot(x='age', y='fare', data=df)

# Add title and labels
plt.title('Scatter Plot of Age vs. Fare in Titanic Dataset')
plt.xlabel('Age')
plt.ylabel('Fare')

# Show the plot
plt.show()
```

### B. Update the scatter plot with additional information

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))  # Adjust the figure size for better readability
sns.scatterplot(x='age', y='fare', marker="D", color = 'red' , data=df)

# Add title and labels
plt.title('Scatter Plot of Age vs. Fare in Titanic Dataset')
plt.xlabel('Age')
plt.ylabel('Fare')

# Show the plot
plt.show()
```

### C. Create multiple visualizations using different types of markers

```python
# Define the columns to plot
columns_to_plot = ['age', 'fare']

# Define the marker styles
all_poss = ['.', 'o', 'v', '^', '>', '<', 's', 'p', '*', 'h', 'H', 'D', 'd', '1']

# Define a list of colors.
colors = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00',
          '#ffff33', '#a65628', '#f781bf', '#999999', '#1b9e77',
          '#d95f02', '#7570b3', '#e7298a', '#66a61e', '#a6761d',
          '#666666', '#8da0cb'] # More colors

# Determine the number of rows and columns for the subplot grid
num_plots = len(all_poss)
num_cols = 4  # You can adjust the number of columns as needed
num_rows = (num_plots + num_cols - 1) // num_cols  # Calculate the number of rows

# Create the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(16, 4 * num_rows))  # Adjust figure size

# Iterate through the markers and create a scatter plot for each
for i, marker in enumerate(all_poss):
    # Determine the row and column for the current subplot
    row = i // num_cols
    col = i % num_cols

    # Select the appropriate subplot axis
    if num_rows == 1:
        ax = axes[col]
    else:
        ax = axes[row, col]

    # Use a different color for each scatter plot.
    color = colors[i % len(colors)]  # Cycle through the colors

    # Create the scatter plot
    sns.scatterplot(x='age', y='fare', data=df, marker=marker, color=color, ax=ax)
    ax.set_title(f'Age vs. Fare (Marker: "{marker}")')
    ax.set_xlabel('Age')
    ax.set_ylabel('Fare')

# Remove any unused subplots (if the number of plots is not a multiple of the number of columns)
if num_plots < num_rows * num_cols:
    for j in range(num_plots, num_rows * num_cols):
        row = j // num_cols
        col = j % num_cols
        if num_rows == 1:
            fig.delaxes(axes[col])
        else:
            fig.delaxes(axes[row, col])

# Adjust the layout to prevent overlapping titles/labels
plt.tight_layout()

# Show the plots
plt.show()
```

# 3. How to interpret Scatter plot

Key Features to Look For

Shape:

* Linear: The points tend to form a straight line.
* Positive: The line slopes upwards from left to right (as one variable increases, the other also increases).
* Negative: The line slopes downwards from left to right (as one variable increases, the other decreases).
* Curvilinear: The points form a curved pattern.
* Random: The points show no clear pattern.

Direction:

* As described under "Linear" shape: Positive or Negative.  Curvilinear and Random patterns don't have a single direction.
* Strength: How closely the points follow the pattern.
* Strong: The points are tightly clustered around the line or curve.
* Moderate: The points show a clear pattern, but with some scatter.
* Weak: The points show a pattern, but it's very dispersed.
* No correlation: The points appear randomly scattered.

Outliers:

* Points that fall far away from the general pattern of the data.  Outliers can have a large influence on the relationship.

Clusters:

* Groups of points that are close together, separated from other groups. Clusters might indicate different subgroups within the data.

Gaps:

* Areas in the plot where there are few or no data points. Gaps can highlight ranges of values that are infrequent.
