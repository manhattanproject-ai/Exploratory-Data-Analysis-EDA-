# 2. Create waterfall charts

### A. Create the waterfall chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the Waterfall Chart
approval_cost = 900
ai_library_screening = -135
automated_preclinical = -45
probability_improvement = -280
precision_therapy_cost = 440

# Calculate cumulative values
cumulative = np.array([
    approval_cost,
    approval_cost + ai_library_screening,
    approval_cost + ai_library_screening + automated_preclinical,
    approval_cost + ai_library_screening + automated_preclinical + probability_improvement,
    precision_therapy_cost
])

# Define labels and colors
labels = [
    'Approval\nCost',
    'AI-Enabled\nLibrary Screening',
    'Automated\nPreclinical\nWorkflows',
    'Savings From\nProbability Of\nApproval\nImprovement',
    'Precision\nTherapy\nApproval Cost'
]
colors = ['#A080FF', '#FF69B4', '#FF69B4', '#FF69B4', '#20B2AA']

# --- Matplotlib Code ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('#f0f0f0') # Match background color

# Create the waterfall chart
ax.bar(labels[0], cumulative[0], color=colors[0], width=0.6)
for i in range(1, len(cumulative) - 1):
    ax.bar(labels[i], cumulative[i] - cumulative[i-1], bottom=cumulative[i-1], color=colors[i], width=0.6)
ax.bar(labels[-1], cumulative[-1], color=colors[-1], width=0.6)

# Add value labels
for i, val in enumerate(cumulative):
    if i == 0:
        ax.text(i, val + 10, f'${val}mm', ha='center', va='bottom', fontsize=10)
    elif i < len(cumulative) - 1:
        change = cumulative[i] - cumulative[i-1]
        ax.text(i, cumulative[i] + np.sign(change) * 10, f'${change}mm', ha='center', va='bottom' if change > 0 else 'top', fontsize=10)
    else:
        ax.text(i, val + 10, f'${val}mm', ha='center', va='bottom', fontsize=10)

# Set y-axis formatting
ax.set_ylabel('Average Cost\nPer Approval ($MM)', fontsize=12)
ax.set_yticks(np.arange(0, 1100, 100))
ax.set_yticklabels([f'${i}' for i in range(0, 1100, 100)], fontsize=10)
ax.set_ylim(0, 1000)

# Set x-axis formatting
ax.set_xticks(np.arange(len(labels)))
ax.set_xticklabels(labels, rotation=0, ha='center', fontsize=10)

# Title
ax.set_title('Forecasted Reduction In Cost/Approval Attributable To AI/Automation', fontsize=14)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_linewidth(0.5)
ax.spines['bottom'].set_color('black')

ax.grid(visible=None)

plt.tight_layout()
plt.show()
```
