# 1. Import necessary dependencies

```python
# libraries & dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="darkgrid")
```

# 2. Create the dataset

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.shape
```

# 3. Box plot with jitters visualization for numerical vs categorical features

### A. Box plot with jitters for species vs sepal_length

```python
# Create the box plot with jitters for sepal_length by species
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal_length', data=iris, palette="viridis")
sns.stripplot(x='species', y='sepal_length', data=iris, color='black', jitter=0.2)
plt.title("sepal_length by species with Jitters")
plt.xlabel("species")
plt.ylabel("sepal_length")
plt.show()
```

### B. Box plot with jitters for all numerical features with categorical variable

```python
# --- Box Plots with Jitters for Numerical vs Categorical Variables ---

numerical_features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
categorical_feature = 'species'

num_plots = len(numerical_features)
rows = (num_plots + 2) // 3
cols = min(num_plots, 2)

fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
fig.suptitle("Box Plots of Iris Features by Species with Jitters", fontsize=16, y=1.02)

axes = axes.flatten()

for i, numerical_col in enumerate(numerical_features):
    if i < len(axes):
        ax = axes[i]
        sns.boxplot(x=categorical_feature, y=numerical_col, data=iris, ax=ax, palette="viridis")
        sns.stripplot(x=categorical_feature, y=numerical_col, data=iris, color='black', jitter=0.2, ax=ax)
        ax.set_title(f"{numerical_col} by {categorical_feature}")
        ax.set_xlabel(categorical_feature)
        ax.set_ylabel(numerical_col)

# Remove any unused subplots
if num_plots < len(axes):
    for j in range(num_plots, len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Interpreting the Attached Box Plots with Jitters:

1. Sepal Length by Species:

* Setosa: The box plot shows the shortest median sepal length with a relatively small IQR. The jittered points are tightly clustered, indicating low variability.
* Versicolor: The median sepal length is intermediate, and the IQR is larger than setosa's. The jittered points are more spread out, showing greater variability.
* Virginica: The box plot indicates the longest median sepal length with the largest IQR. The jittered points are the most dispersed, suggesting the highest variability in sepal length among the three species.
* Jitters: The individual black points clearly illustrate the density of measurements for each species. You can see where most of the data points are concentrated and get a better sense of the distribution's shape.

2. Sepal Width by Species:

* Setosa: The box plot shows the highest median sepal width. The jittered points are somewhat spread, indicating moderate variability.
* Versicolor: The median sepal width is lower than setosa's, and the IQR is smaller. The jittered points are more tightly clustered than setosa's.
* Virginica: The median sepal width is similar to versicolor's, but the IQR is slightly larger. The jittered points show a moderate spread.
* Jitters: The jittered points help visualize the overlap in sepal width between the species, particularly between versicolor and virginica. You can also see the presence of a few outliers for each species.

3. Petal Length by Species:

* Setosa: The box plot shows the shortest median petal length with a very small IQR. The jittered points are tightly packed, indicating very low variability.
* Versicolor: The median petal length is significantly longer than setosa's, and the IQR is larger. The jittered points are more spread out.
* Virginica: The box plot indicates the longest median petal length with the largest IQR. The jittered points are the most dispersed.
* Jitters: The jittered points clearly demonstrate the distinct separation in petal length, especially between setosa and the other two species. The increasing spread from setosa to virginica is also evident.

4. Petal Width by Species:

* Setosa: The box plot shows the smallest median petal width with a very small IQR. The jittered points are tightly clustered.
* Versicolor: The median petal width is significantly larger than setosa's, and the IQR is larger. The jittered points are more spread out.
* Virginica: The box plot indicates the largest median petal width with the largest IQR. The jittered points are the most dispersed.
* Jitters: Similar to petal length, the jittered points highlight the clear separation in petal width between the species. The increasing variability from setosa to virginica is also apparent.

Overall Interpretation:

The box plots with jitters provide a comprehensive view of the Iris dataset features by species:

* Central Tendency and Spread: The boxes and whiskers summarize the median and variability.
* Individual Data Points: The jittered black points show the actual measurements and their distribution, revealing density and potential multimodal patterns (though less evident here).
* Overlap: The jittered points help visualize the degree of overlap between the species for each feature.
* Outliers: Individual points lying outside the whiskers are clearly visible.

# 4. Create new dataset

```python
wine_data_red= "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
wine_df = pd.read_csv(wine_data_red, sep=';')
```

# 5. Box plot with jitters visualization for numerical vs categorical features

### A. Box plot with jitters for quality vs citric acid

```python
# Create the box plot with jitters for citric acid by quality
plt.figure(figsize=(20, 6))
sns.boxplot(x='quality', y='citric acid', data=wine_df, palette="viridis")
sns.stripplot(x='quality', y='citric acid', data=wine_df, color='black', jitter=0.2)
plt.title("citric acid by quality with Jitters")
plt.xlabel("quality")
plt.ylabel("citric acid")
plt.show()
```

### B. Box plot with jitters for all numerical features with categorical variable

```python
# --- Box Plots with Jitters for Numerical vs Categorical Variables ---

numerical_features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
                      'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
                      'pH', 'sulphates', 'alcohol']
categorical_feature = 'quality'

num_plots = len(numerical_features)
rows = (num_plots + 2) // 3
cols = min(num_plots, 3)

fig, axes = plt.subplots(rows, cols, figsize=(20, 5 * rows))
fig.suptitle("Box Plots of Wine Features by quality with Jitters", fontsize=16, y=1.02)

axes = axes.flatten()

for i, numerical_col in enumerate(numerical_features):
    if i < len(axes):
        ax = axes[i]
        sns.boxplot(x=categorical_feature, y=numerical_col, data=wine_df, ax=ax, palette="viridis")
        sns.stripplot(x=categorical_feature, y=numerical_col, data=wine_df, color='black', jitter=0.2, ax=ax)
        ax.set_title(f"{numerical_col} by {categorical_feature}")
        ax.set_xlabel(categorical_feature)
        ax.set_ylabel(numerical_col)

# Remove any unused subplots
if num_plots < len(axes):
    for j in range(num_plots, len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Interpreting Individual Box Plots with Jitters:

* Fixed Acidity by Quality: The median fixed acidity appears relatively stable across quality ratings. The jittered points show a wide spread at each quality level, indicating considerable variability. There are also several outliers, especially at lower quality levels.
* Volatile Acidity by Quality: There's a clear negative trend. As quality increases, the median volatile acidity tends to decrease, and the bulk of the jittered points shift towards lower values. Higher quality wines show a denser cluster of points at lower volatile acidity.
* Citric Acid by Quality: The median citric acid generally shows a positive trend with increasing quality, although it might plateau at higher quality. The jittered points are more concentrated at higher citric acid levels for higher quality wines.
* Residual Sugar by Quality: The relationship is complex. The median doesn't show a strong linear trend. The jittered points reveal a wide range of residual sugar at all quality levels, with some high values even in high-quality wines.
* Chlorides by Quality: A negative trend is evident. Higher quality wines tend to have a denser cluster of jittered points at lower chloride levels. The spread decreases somewhat at higher qualities.
* Free Sulfur Dioxide by Quality: The median free sulfur dioxide is relatively stable. The jittered points show a considerable spread at each quality level, with some wines having very high free sulfur dioxide regardless of quality.
* Total Sulfur Dioxide by Quality: There's a slight negative trend in the median total sulfur dioxide as quality increases, but it's not very strong. The jittered points show a wide distribution, with many outliers, especially at lower quality levels.
* Density by Quality: The median density is very consistent across quality ratings. The jittered points form relatively tight clusters around the median, indicating low variability for density at each quality level.
* pH by Quality: The median pH shows a slight positive trend with increasing quality. The jittered points are somewhat spread out at each quality level, indicating moderate variability.
* Sulphates by Quality: A positive trend is noticeable. Higher quality wines tend to have a denser cluster of jittered points at higher sulphate levels. The spread also appears to decrease slightly at higher qualities.
* Alcohol by Quality: There's a clear positive correlation. As quality increases, the median alcohol content is generally higher, and the jittered points are concentrated at progressively higher alcohol percentages.

Key Insights from Adding Jitters:

* Data Density: The jittered points provide a much clearer picture of where the majority of the data lies within each quality category, revealing potential skewness or multimodal distributions that might be hidden by the box plots alone.
* Overlap: The jitter helps visualize the overlap in the feature values between different quality ratings. For example, while the median volatile acidity is lower for higher quality wines, there are still some lower quality wines with similar volatile acidity levels.
* Outlier Distribution: The jittered points make it easier to see the frequency and distribution of outliers at each quality level.

In summary, adding jitters to the box plots enhances the visualization by showing the underlying distribution of individual wine samples. This provides a more nuanced understanding of how each chemical feature relates to the perceived quality of wine, highlighting the central tendency, spread, density, and the presence of outliers within each quality group.
