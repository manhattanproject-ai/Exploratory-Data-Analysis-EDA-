3. Create Clustered bar chart overtime


# 1. Import necessary dependencies

```python
# libraries & dataset
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid")
```

# 2. Create the dataset

```python
# Approximate data for the number of months to reach 100 million users
months = np.linspace(0, 5, 50)

data = {
    'Months': months,
    'ChatGPT': np.clip(100 * (months / 2)**3, 0, 100),
    'Instagram': np.clip(100 * (months / 2.5)**2, 0, 100),
    'TikTok': np.clip(100 * (months / 1.8)**2, 0, 100),
    'WeChat': np.clip(100 * (months / 2.2)**2, 0, 100),
    'YouTube': np.clip(100 * (months / 3.5)**2, 0, 100),
    'Facebook': np.clip(100 * (months / 4.5)**2, 0, 100)
}
```

```python
# Create the Pandas DataFrame
df = pd.DataFrame(data)
df
```

# 3. Create clustered bar chart overtime

```python
fig, ax = plt.subplots(figsize=(12, 7))

ax.plot(df['Months'], df['ChatGPT'], color='#A020F0', linestyle='-', label='ChatGPT', linewidth=2)
ax.plot(df['Months'], df['Instagram'], color='#343A40', linestyle='--', label='Instagram', linewidth=2)
ax.plot(df['Months'], df['TikTok'], color='#17A2B8', linestyle='--', label='TikTok', linewidth=2)
ax.plot(df['Months'], df['WeChat'], color='#6C757D', linestyle='--', label='WeChat', linewidth=2)
ax.plot(df['Months'], df['YouTube'], color='#FFC107', linestyle='--', label='YouTube', linewidth=2)
ax.plot(df['Months'], df['Facebook'], color='#28A745', linestyle='--', label='Facebook', linewidth=2)

# Set y-axis formatting
ax.set_ylabel('Monthly Active Users\n(Millions)', fontsize=12, rotation=90, ha='right')
ax.set_yticks(np.arange(0, 101, 10))
ax.set_yticklabels([f'{i}' for i in range(0, 101, 10)], fontsize=10)
ax.set_ylim(0, 105)

# Set x-axis formatting
ax.set_xlabel('Years', fontsize=12)
ax.set_xticks(np.arange(0, 6, 1))
ax.set_xticklabels([f'{i}' for i in range(0, 6, 1)], fontsize=10)
ax.set_xlim(0, 5)

# Set title
ax.set_title('ChatGPT Users Hit 100 Million Users In Two Months', fontsize=14)

# Add legend
ax.legend(loc='upper left', fontsize=10, frameon=False, ncol=3)

# Add annotation
ax.text(0, -10, '*values between 0 and 100 million users are estimates', fontsize=9, ha='left', va='top', color='grey')

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_linewidth(0.5)
ax.spines['bottom'].set_color('black')

ax.grid(visible=None)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
