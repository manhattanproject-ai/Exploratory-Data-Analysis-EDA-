# 1. Import necessary dependencies

```python
# libraries & dataset
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

sns.set_theme(style="darkgrid")
```

# 2. Create the dataset

```python
# Data for the chart (in Trillions of US Dollars)
data = {
    'Year': [2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030],
    'Direct': [0.6, 0.7, 0.8, 1.1, 1.1, 1.2, 1.3, 1.6, 2.0, 2.3, 2.7, 3.3, 4.0],
    'Indirect': [0.3, 0.4, 0.5, 0.4, 0.5, 0.6, 0.7, 0.8, 0.8, 0.9, 1.0, 1.1, 1.2]
}
```

```python
# Create the Pandas DataFrame
df = pd.DataFrame(data)
df
```

# 3. Create stacked bar chart overtime

```python
fig, ax = plt.subplots(figsize=(15, 8))

bar_width = 0.7

# Create the stacked bar chart
ax.bar(df['Year'], df['Direct'], color='#9370DB', label='Direct', width=bar_width)
ax.bar(df['Year'], df['Indirect'], bottom=df['Direct'], color='#36454F', label='Indirect', width=bar_width)

# Add value labels inside the stacked bars
for i in range(len(df)):
    # Label for Direct revenue
    ax.text(df['Year'][i], df['Direct'][i] / 2, f"${df['Direct'][i]:.1f}T",
            ha='center', va='center', color='white', fontsize=9)
    # Label for Indirect revenue
    ax.text(df['Year'][i], df['Direct'][i] + df['Indirect'][i] / 2, f"${df['Indirect'][i]:.1f}T",
            ha='center', va='center', color='white', fontsize=9)

# Set x-axis ticks and labels
ax.set_xticks(df['Year'])
# ax.set_xticklabels([str(year)[2:] if year != df['Year'][0] else str(year) for year in df['Year']], fontsize=10)

# Set y-axis ticks and labels
ax.set_yticks(np.arange(0, 7, 1))
ax.set_yticklabels(['$-', '$1', '$2', '$3', '$4', '$5', '$6'], fontsize=10)
ax.set_ylabel('$ Trillions', fontsize=12)

# Set title
ax.set_title('Global Digital Platform Revenue', fontsize=14)

# Add legend
ax.legend(loc='upper left', fontsize=10, frameon=False, ncol=2)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_linewidth(0.5)
ax.spines['left'].set_position(('outward', 10))
ax.spines['bottom'].set_position(('outward', 5))

# Add grid lines (optional)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```
