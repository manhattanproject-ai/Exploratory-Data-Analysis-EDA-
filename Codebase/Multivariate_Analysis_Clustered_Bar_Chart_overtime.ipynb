# 1. Import necessary dependencies

```python
# libraries & dataset
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid")
```

# 2. Create the dataset

```python
# Data for Annualized Returns Across Major Asset Classes
data = {
    'Time Period': ['Last 7 Years', 'Last 6 Years', 'Last 5 Years', 'Last 4 Years', 'Last 3 Years'],
    'Bitcoin': [72, 63, 62, 56, 12],
    'Gold': [8, 9, 9, 8, 7],
    'Commodities': [3, 4, 5, 6, 2],
    'Real Estate': [4, 5, 6, 7, 13],
    'Bonds': [13, 15, 16, 15, 18],
    'Equities': [7, 6, 7, 5, 10],
    'Emerging Markets': [2, 3, 4, -2, -3]
}
```

```python
# Create the Pandas DataFrame
df = pd.DataFrame(data)
df
```

# 3. Create clustered bar chart overtime

```python
fig, ax = plt.subplots(figsize=(15, 7))

bar_width = 0.10
x = np.arange(len(df.index))

ax.bar(x - 3*bar_width/2, df['Bitcoin'], color='#FFDA63', width=bar_width, label='Bitcoin')
ax.bar(x - bar_width/2, df['Gold'], color='#D4AC0D', width=bar_width, label='Gold')
ax.bar(x + bar_width/2, df['Commodities'], color='#5DADE2', width=bar_width, label='Commodities')
ax.bar(x + 3*bar_width/2, df['Real Estate'], color='#3498DB', width=bar_width, label='Real Estate')
ax.bar(x + 5*bar_width/2, df['Bonds'], color='#9B59B6', width=bar_width, label='Bonds')
ax.bar(x + 7*bar_width/2, df['Equities'], color='#E74C3C', width=bar_width, label='Equities')
ax.bar(x + 9*bar_width/2, df['Emerging Markets'], color='#A04000', width=bar_width, label='Emerging Markets')

# Set y-axis formatting
ax.set_ylabel('CAGR (%)', fontsize=12)
ax.set_yticks(np.arange(-10, 90, 10))
ax.set_yticklabels([f'{i}%' for i in range(-10, 90, 10)], fontsize=10)
ax.set_ylim(-10, 80)

# Set x-axis formatting
ax.set_xlabel('', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(df['Time Period'], fontsize=10)
# ax.set_xticklabels(df.index, fontsize=10)

# Add horizontal lines
ax.axhline(y=44, color='#F1C40F', linestyle='--', linewidth=1, alpha=0.7, label='Average Bitcoin CAGR: ~44%')
ax.axhline(y=5.7, color='black', linestyle='--', linewidth=1, alpha=0.7, label='Average Asset Class CAGR: 5.7%')

# Set title
ax.set_title('Annualized Returns Across Major Asset Classes', fontsize=14)

# Add legend
ax.legend(loc='upper left', fontsize=10, ncol=7, frameon=False)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_linewidth(0.5)
ax.spines['bottom'].set_color('black')

ax.grid(visible=None)

# Adjust layout
plt.tight_layout()
plt.show()
```
