# 1. Import necessary dependencies

```python
# libraries & dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="darkgrid")
```

# 2. Create the dataset

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.shape
```

# 3. Violin plot visualization for numerical vs categorical features

### A. Violin plot for species vs sepal_length

```python
# Create the violin plot for sepal_length by species
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='sepal_length', data=iris, palette="viridis")
plt.title("sepal_length by species")
plt.xlabel("species")
plt.ylabel("sepal_length")
plt.show()
```

### B. Violin plot for all numerical features with categorical variable

```python
# --- violin Plots for Numerical vs Categorical Variables ---

numerical_features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
categorical_feature = 'species'

num_plots = len(numerical_features)
rows = (num_plots + 2) // 3
cols = min(num_plots, 2)

fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
fig.suptitle("Violin Plots of Iris Features by Species", fontsize=16, y=1.02)

axes = axes.flatten()

for i, numerical_col in enumerate(numerical_features):
    if i < len(axes):
        ax = axes[i]
        sns.violinplot(x=categorical_feature, y=numerical_col, data=iris, ax=ax, palette="viridis")
        ax.set_title(f"{numerical_col} by {categorical_feature}")
        ax.set_xlabel(categorical_feature)
        ax.set_ylabel(numerical_col)

# Remove any unused subplots
if num_plots < len(axes):
    for j in range(num_plots, len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Interpreting the Attached Violin Plots:

1. Sepal Length by Species:

* Setosa: Shows a unimodal distribution (one main peak) centered around a shorter sepal length. The violin is relatively narrow, indicating less spread compared to the other species.
* Versicolor: Exhibits a wider, somewhat flatter distribution, suggesting more variability in sepal length. It appears slightly bimodal or has a broader peak.
* Virginica: Displays a distribution with a peak at a longer sepal length. It also shows considerable spread, with a slightly skewed shape towards lower values.
* Comparison: Setosa has the shortest sepal lengths with the least variability. Virginica has the longest sepal lengths with more variability. Versicolor falls in between with a broader distribution.

2. Sepal Width by Species:

* Setosa: Shows a distribution centered around a larger sepal width with moderate spread. It appears somewhat bimodal or has a wider peak.
* Versicolor: Exhibits a distribution with a peak at a smaller sepal width than setosa, with less spread than setosa.
* Virginica: Displays a distribution centered around a sepal width similar to versicolor but with a wider spread and a slightly flatter shape.
* Comparison: Setosa tends to have wider sepals, while versicolor and virginica have narrower sepals with more overlap in their distributions.

3. Petal Length by Species:

* Setosa: Shows a very narrow, unimodal distribution centered around a short petal length, indicating very low variability.
* Versicolor: Exhibits a wider, unimodal distribution centered around a petal length significantly longer than setosa and shorter than virginica.
* Virginica: Displays a broader, unimodal distribution centered around the longest petal length, with more spread than versicolor.
* Comparison: Petal length is a strong distinguishing feature. Setosa has distinctly short petals. Versicolor and virginica have longer petals with relatively less overlap compared to sepal length.

4. Petal Width by Species:

* Setosa: Shows a very narrow, unimodal distribution centered around a small petal width, indicating very low variability.
* Versicolor: Exhibits a wider, unimodal distribution centered around a petal width significantly larger than setosa and smaller than virginica.
* Virginica: Displays a broader, unimodal distribution centered around the largest petal width, with more spread than versicolor.
* Comparison: Similar to petal length, petal width effectively separates the species. Setosa has distinctly narrow petals. Versicolor and virginica have wider petals with minimal overlap.

Overall Interpretation:

These violin plots provide a more detailed view of the distribution of each Iris flower feature compared to simple box plots. They reinforce the following observations:

* Petal length and petal width are the most effective features for distinguishing the three Iris species, showing clear separation in their distributions.
* Sepal length also helps in distinguishing the species, with setosa having shorter sepals and virginica longer sepals. Versicolor overlaps more with both.
* Sepal width shows the most overlap between the species, making it a less reliable feature for classification. Setosa tends to have wider sepals.

# 4. Create new dataset

```python
wine_data_red= "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
wine_df = pd.read_csv(wine_data_red, sep=';')
```

# 5. Violin plot visualization for numerical vs categorical features

### A. Violin plot for quality vs citric acid

```python
# Create the violin plot for citric acid by quality
plt.figure(figsize=(15, 6))
sns.violinplot(x='quality', y='citric acid', data=wine_df, palette="viridis")
plt.title("citric acid by quality")
plt.xlabel("quality")
plt.ylabel("citric acid")
plt.show()
```

### B. Violin plot for all numerical features with categorical variable

```python
# --- violin Plots for Numerical vs Categorical Variables ---

numerical_features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']
categorical_feature = 'quality'

num_plots = len(numerical_features)
rows = (num_plots + 2) // 3
cols = min(num_plots, 4)

fig, axes = plt.subplots(rows, cols, figsize=(20, 5 * rows))
fig.suptitle("Violin Plots of Wine Features by quality", fontsize=16, y=1.02)

axes = axes.flatten()

for i, numerical_col in enumerate(numerical_features):
    if i < len(axes):
        ax = axes[i]
        sns.violinplot(x=categorical_feature, y=numerical_col, data=wine_df, ax=ax, palette="viridis")
        ax.set_title(f"{numerical_col} by {categorical_feature}")
        ax.set_xlabel(categorical_feature)
        ax.set_ylabel(numerical_col)

# Remove any unused subplots
if num_plots < len(axes):
    for j in range(num_plots, len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Interpretation of the violin plots

Interpreting Individual Violin Plots:

* Fixed Acidity by Quality: The distributions appear relatively similar across quality ratings, with a slight tendency for higher quality wines to have a slightly higher median fixed acidity. The spread is also comparable.
* Volatile Acidity by Quality: There's a clear negative trend. As quality increases, the median volatile acidity tends to decrease, and the bulk of the distribution shifts towards lower values. Higher quality wines show a much narrower distribution at lower volatile acidity levels.
* Citric Acid by Quality: The median citric acid generally increases with quality, although it seems to plateau at higher quality levels. The distribution for higher quality wines also tends to be skewed towards higher citric acid values.
* Residual Sugar by Quality: The relationship is complex. Lower quality wines show a wider spread of residual sugar. The median doesn't show a strong linear trend. Some high-quality wines have very low residual sugar, while others might have higher levels. The distributions are often skewed.
* Chlorides by Quality: A negative trend is evident. Higher quality wines tend to have lower median chloride levels, and the distribution is concentrated at lower chloride values.
* Free Sulfur Dioxide by Quality: The medians are relatively stable across quality ratings. However, the spread seems to decrease slightly for higher quality wines, suggesting a more consistent range of free sulfur dioxide.
* Total Sulfur Dioxide by Quality: There's a slight negative trend in the median total sulfur dioxide as quality increases, although it's not very pronounced. The distributions are generally right-skewed.
* Density by Quality: The median density is quite consistent across quality ratings, with a very slight decrease at the highest quality (8). The distributions are narrow, indicating relatively little variability in density for each quality level.
* pH by Quality: The median pH shows a slight positive trend with increasing quality, suggesting higher quality wines tend to be slightly less acidic (higher pH). The distributions are relatively consistent in shape and spread.
* Sulphates by Quality: A positive trend is noticeable. Higher quality wines tend to have higher median sulphate levels, and the distributions shift towards higher sulphate values.
* Alcohol by Quality: There's a clear positive correlation. As quality increases, the median alcohol content is generally higher, and the distributions are centered at progressively higher alcohol percentages.

Summary of Trends:

These violin plots visually reinforce some key relationships with wine quality:

* Negative Indicators of Quality: Higher volatile acidity and higher chlorides are generally associated with lower quality.
* Positive Indicators of Quality: Higher citric acid, higher sulphates, and higher alcohol content tend to be associated with higher quality.
* Less Clear or Complex Relationships: Fixed acidity, residual sugar, free sulfur dioxide, total sulfur dioxide, and pH show weaker or more nuanced relationships with quality.
