# 1. Import necessary dependencies

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
```

# 2. Import the required dataset

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

# 3. Visualize different columns

### A. Drawing a simple Boxplot with default parameters

```python
# Make default box plot
sns.boxplot(y=df["sepal_length"])
plt.show()
```

### B. Improving the graph with different parameters

* Specify the data and the x-variable.
* Set the color of the box.
* Set the caps for the whiskers.
* Set the Whisker length.
* Show the outliers
* Add a title and axis labels.

```python
# Create the box plot with more informative details
sns.boxplot(
    y=df["sepal_width"],
    color="red",
    showcaps=True,  # Show the caps on the whiskers
    whis=1.5,        # Set the whisker length (default is 1.5 * IQR)
    showfliers=True,  # Show the outliers
)
plt.title("Sepal Length Distribution with Quartiles, Whiskers, and Outliers")
plt.ylabel("Sepal Length (cm)")
plt.xlabel("")
plt.show()
```

### C. Visualize multiple Boxplots together

1. Define the columns to plot ( update based on your scenario )
2. Define the colors of each density plot ( update based on your choice/preference )
3. we created a 2 * 2 matrix with each figure as (14 , 10 ) , you can update each of them based on preference


```python
# Define the columns to plot
columns_to_plot = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Define colors for each boxplot
colors = ['red', 'green', 'orange', 'coral']  # Add more colors if needed

# Create a 2x2 subplot grid with a fixed figure size
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Iterate through the columns and plot each one in a separate subplot
for i, col in enumerate(columns_to_plot):
    # Determine the row and column for the current subplot
    row = i // 2
    col_num = i % 2

    # Select the appropriate subplot axis
    ax = axes[row, col_num]

    # Create the boxplot, using a different color for each
    sns.boxplot(data=df, y=df[col], color=colors[i],  showcaps=True, whis=1.5, showfliers=True, ax=ax)

    # Set the title and labels for the subplot
    ax.set_title(f"{col.title()} Distribution")
    ax.set_ylabel(f"{col.title()} (cm)")  # Changed from xlabel to ylabel
    ax.set_xlabel("")
    #ax.legend() #Removed Legend

# Adjust the layout to prevent overlapping titles/labels
plt.tight_layout()

# Display the plots
plt.show()
```

# 4. How to Interpret Boxplot :

Key Features to Look For

Shape:

* Symmetric: In a symmetric distribution, the median is in the center of the box, and the whiskers are roughly the same length.

Skewed:

* Right-skewed (positive skew): The median is closer to the bottom of the box, and the upper whisker is longer.
* Left-skewed (negative skew): The median is closer to the top of the box, and the lower whisker is longer.

Center:

* The median is represented by the line inside the box. It's the middle value of the data.

Spread:

* The box itself shows the interquartile range (IQR), which represents the middle 50% of the data. A taller box indicates a larger spread.
* The whiskers extend to show the range of the data, excluding outliers. Longer whiskers also indicate a larger spread.

Outliers:

Points outside the whiskers are considered outliers. They are data points that are significantly different from the rest of the data.

Gaps:

Box plots don't directly show gaps in the data.

# 5. Create boxplot with jitters

### A. Drawing a boxplot with jitter

To add jitter to a box plot and visualize the distribution of individual data points, you can combine a box plot with a scatter plot.

```python
# Create the box plot
sns.boxplot(y=df["sepal_length"])

# Add a scatter plot with jitter
sns.stripplot(y=df["sepal_length"], color="black", alpha=0.5, jitter=0.2)  # Adjust jitter value as needed

plt.title("Sepal Length Distribution with Jitter")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

### C. Visualize multiple box plots together with jitters

* Define the columns to plot ( update based on your scenario )
* Define the colors of each cumulative density plot ( update based on your choice/preference )
* we created a 2 * 2 matrix with each figure as (14 , 10 ) , you can update each of them based on preference

```python
# Define the columns to plot
columns_to_plot = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Define colors for each boxplot
colors = ['red', 'green', 'orange', 'coral']  # Add more colors if needed

# Create a 2x2 subplot grid with a fixed figure size
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Iterate through the columns and plot each one in a separate subplot
for i, col in enumerate(columns_to_plot):
    # Determine the row and column for the current subplot
    row = i // 2
    col_num = i % 2

    # Select the appropriate subplot axis
    ax = axes[row, col_num]

    # Create the box plot
    sns.boxplot(y=df[col], color=colors[i], showcaps=True, whis=1.5, showfliers=True, ax=ax)

    # Add a scatter plot with jitter
    sns.stripplot(y=df[col], color="black", alpha=0.5, jitter=0.2, ax=ax)  # Adjust jitter value as needed

    # Set the title and labels for the subplot
    ax.set_title(f"{col.title()} Distribution with Jitter")
    ax.set_ylabel(f"{col.title()} (cm)")
    ax.set_xlabel("")

# Adjust the layout to prevent overlapping titles/labels
plt.tight_layout()

# Display the plots
plt.show()
```

# 6. How to Interpret Boxplot with jitter :

Key Features to Look For

Shape:

* Symmetric: In a symmetric distribution, the median is in the center of the box, the whiskers are roughly the same length, and the jittered points are distributed relatively evenly on both sides of the box.

Skewed:

* Right-skewed (positive skew): The median is closer to the bottom of the box, the upper whisker is longer, and the jittered points are more concentrated towards the lower values.
* Left-skewed (negative skew): The median is closer to the top of the box, the lower whisker is longer, and the jittered points are more concentrated towards the higher values.

Center:

* The median is represented by the line inside the box. It's the middle value of the data. The jittered points help visualize the density of data around the median.

Spread:

* The box itself shows the interquartile range (IQR), which represents the middle 50% of the data. A taller box indicates a larger spread. The jittered points show how the data is distributed within the IQR.
* The whiskers extend to show the range of the data, excluding outliers. Longer whiskers indicate a larger spread. The jittered points show the distribution of data points within the range covered by the whiskers.

Outliers:

* Points outside the whiskers are considered outliers and are shown individually. The jittered points make it easier to see how many outliers there are and their values.

Gaps:

* Box plots don't directly show gaps in the data. However, the jittered points can reveal areas where data is sparse, even if the box and whiskers don't explicitly show a gap.

Additional information from Jitter:

* Density/Clusters: The jittered points show the density of the data. Densely packed points indicate areas where many data points have similar values.
* Individual Data Points: Jitter allows you to see each data point, which can be helpful for understanding the distribution and identifying patterns that might not be apparent from the box plot alone.
