# 1. Import necessary dependencies

```python
# libraries & dataset
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd


sns.set_theme(style="darkgrid")
```

# 2. Create the dataset

```python
# 1. Create a dataset with around 10 numerical variables
# Using a dictionary to create the data, then converting to a DataFrame
data = {
    'StockPrice': np.random.normal(100, 15, 100),  # Example stock price
    'EPS': np.random.normal(5, 2, 100),             # Earnings per share
    'PE_Ratio': np.random.normal(20, 5, 100),        # Price-to-earnings ratio
    'DebtToEquity': np.random.uniform(0, 2, 100),   # Debt-to-equity ratio
    'ROA': np.random.normal(0.05, 0.02, 100),       # Return on assets
    'ROE': np.random.normal(0.10, 0.05, 100),      # Return on equity
    'MarketCap': np.random.lognormal(10, 2, 100),   # Market capitalization
    'RevenueGrowth': np.random.normal(0.08, 0.03, 100), # Revenue growth rate
    'DividendYield': np.random.uniform(0, 0.05, 100), # Dividend yield
    'Beta': np.random.normal(1, 0.3, 100)           # Beta coefficient
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.shape
```

# 3. Jointplot visualization for the numerical features

### A. Jointplot with Hexbin

```python
# Create a joint plot
plt.figure(figsize=(8, 6))  # Adjust the figure size for better readability
sns.jointplot(x='StockPrice', y='EPS', data=df, kind='hex')

# Add title (jointplot handles axis labels)
plt.suptitle('Joint Plot of StockPrice vs. EPS in Stock dataset', y=1.02)

# Show the plot
plt.show();
```

### B. Joint plot with scatter

```python
# Create a Jointplot plot
plt.figure(figsize=(8, 6))  # Adjust the figure size for better readability
sns.jointplot(x='StockPrice', y='PE_Ratio', data=df, kind='scatter')

# Add title (jointplot handles axis labels)
plt.suptitle('Joint Plot of StockPrice vs. PE_Ratio in Stock dataset', y=1.02)

# Show the plot
plt.show();
```

### C. Change the hexagon size

```python
# Change the number of bins (hexagon size)
plt.figure(figsize=(8, 6))
sns.jointplot(x='StockPrice', y='ROE', data=df, kind='hex', gridsize=20) # Try different values for gridsize
plt.suptitle('Joint Plot with Custom Grid Size', y=1.02)
plt.show()
```

```python
# Change the color map of the hexbins
plt.figure(figsize=(8, 6))
sns.jointplot(x='StockPrice', y='RevenueGrowth', data=df, kind='hex', cmap='RdBu') # Try other colormaps like 'plasma', 'magma', 'cividis'
plt.suptitle('Joint Plot with Custom Color Map', y=1.02)
plt.show();
```

```python
# Add marginal distributions (histograms or KDEs)
plt.figure(figsize=(8, 6))
sns.jointplot(x='StockPrice', y='Beta', data=df, kind='hex', marginal_kws=dict(bins=15, color='red'))
plt.suptitle('Joint Plot with Marginal Histograms', y=1.02)
plt.show();
```

# 4. How to interpret Joint plot

Key Features to Look For:

1. Central Bivariate Plot (Relationship between the two variables):

Pattern/Shape:

* Linear: Points (in a scatter plot or implied by the density in a hexbin or KDE plot) tend to follow a straight line.
  * Positive Linear: As one variable increases, the other tends to increase.
  * Negative Linear: As one variable increases, the other tends to decrease.
* Non-linear: Points follow a curved pattern, indicating a more complex relationship.
* Clustering: Data points form distinct groups or clusters.
* No Clear Pattern: Points are scattered randomly, suggesting a weak or no linear relationship.
* Center (of the relationship):
  * For scatter plots, visually assess the general central tendency of the point cloud.
  * For hexbin plots, the darkest hexagons indicate the areas of highest concentration.
  * For KDE plots, the innermost contours represent the highest density.
  * For regression plots, the regression line shows the central tendency of the linear relationship.
* Spread (of the relationship):
  * A wide scattering of points (or a broad area of density in hexbin/KDE) indicates a weaker relationship or more variability.
  * Points tightly clustered around a line or a dense area suggest a stronger relationship.
* Outliers (in the bivariate space): Points that are far away from the main cluster or pattern in the central plot might be bivariate outliers.
* Gaps (in the bivariate space): Regions in the central plot with very few or no data points might indicate gaps or a lack of observations for those combinations of the two variables.

2. Top Marginal Plot (Distribution of the X-axis Variable):

Shape:

* Symmetric: The distribution is balanced around a central point (e.g., bell-shaped or normal).
* Skewed:
  * Right-skewed (positive skew): The tail of the distribution extends to the right (higher values).
  * Left-skewed (negative skew): The tail of the distribution extends to the left (lower values).
* Unimodal: One peak in the distribution.
* Bimodal or Multimodal: Multiple peaks, suggesting subgroups within the data.
* Center:
  * Look at the peak of the histogram or KDE to estimate the mode (most frequent value).
  * The central tendency can also be inferred visually.
* Spread:
  * A wider histogram or KDE indicates a larger range and more variability in the x-axis variable.
  * A narrower histogram or KDE indicates a smaller range and less variability.
* Outliers: In a histogram, isolated bars far from the main distribution might suggest outliers. The tails of a KDE can also hint at potential outliers.
* Gaps: Empty bins in a histogram indicate gaps in the observed values of the x-axis variable.

3. Right Marginal Plot (Distribution of the Y-axis Variable):

Interpret the shape, center, spread, outliers, and gaps in the distribution of the y-axis variable in the same way as described for the top marginal plot.
