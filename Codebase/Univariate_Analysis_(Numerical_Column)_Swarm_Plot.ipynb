# 1. Import necessary dependencies

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
```

# 2. Import the required dataset

```python
df = sns.load_dataset("iris")
```

# 3. Visualize different columns

### A. Drawing a simple Swarm plot with default parameters

```python
# Make default Swarm plot
sns.swarmplot(y=df["sepal_length"])
plt.show()
```

### B. Improving the graph with different parameters

* Specify the data and the x-variable.
* Set the color of the points.
* Add a title and axis labels.

```python
sns.swarmplot(
    x=None,  # If you want a horizontal swarm plot, specify the column for the x-axis here.  If y is specified, and x is None, the points are plotted vertically.
    y=df["sepal_width"],  # Column for the y-axis
    color="red",  # Column to determine the color of the points
    data=df,  # DataFrame to use
    order=None,  # Order of categories along the x-axis
    hue_order=None,  # Order of hue categories
    size=8,  # Size of the markers
    palette=None,  # Colors to use for different levels of the hue variable
    ax=None,  #  Matplotlib axes object to plot onto. Otherwise, uses the current active axes.
)

plt.title("Swarm Plot of Sepal Width")
plt.ylabel("Sepal Width (cm)")
plt.xlabel("")
plt.show()
```

### C. Visualize multiple Swarm plot together

1. Define the columns to plot ( update based on your scenario )
2. Define the colors of each Swarm plot ( update based on your choice/preference )
3. we created a 2 * 2 matrix with each figure as (14 , 10 ) , you can update each of them based on preference


```python
# Define the columns to plot
columns_to_plot = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Define colors for each swarm plot
colors = ['red', 'green', 'orange', 'coral']  # Add more colors if needed

# Create a 2x2 subplot grid with a fixed figure size
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Iterate through the columns and plot each one in a separate subplot
for i, col in enumerate(columns_to_plot):
    # Determine the row and column for the current subplot
    row = i // 2
    col_num = i % 2

    # Select the appropriate subplot axis
    ax = axes[row, col_num]

    # Create the swarm plot, using a different color for each
    sns.swarmplot(data=df, y=df[col], color=colors[i], ax=ax)

    # Set the title and labels for the subplot
    ax.set_title(f"{col.title()} Distribution")
    ax.set_ylabel(f"{col.title()} (cm)")  # Changed from xlabel to ylabel
    ax.set_xlabel("")

# Adjust the layout to prevent overlapping titles/labels
plt.tight_layout()

# Display the plots
plt.show()
```

# 4. How to Interpret Swarm plot :

Key Features to Look For

Shape:

* A swarm plot shows the distribution of data points. It's similar to a strip plot, but the points are adjusted along the x-axis (or y-axis if the plotted variable is on the x-axis) so they don't overlap.
* Since a swarm plot displays individual data points, the "shape" you observe is how those points are distributed.
* Symmetric: If the points are distributed evenly on both sides of the center, the distribution is symmetric.

Skewed:

* Right-skewed (positive skew): Most points are concentrated on the left, with fewer points extending to the right.
* Left-skewed (negative skew): Most points are concentrated on the right, with fewer points extending to the left.
* Unimodal/Bimodal/Multimodal: Clusters of points can indicate peaks in the data.

Unimodal: A single dense cluster.

Bimodal: Two dense clusters.

Multimodal: More than two dense clusters.

Center:

* The center of the data can be visually estimated. It's where the points are most concentrated.

Spread:

* The spread is how much the points vary along the axis. A wider spread means the data points have more variability.

Outliers:

* Outliers are easily identifiable as points that are far away from the main cluster of data points.

Gaps:

* Swarm plots can show gaps in the data. Gaps appear as empty areas along the axis where there are no points.
