# 3. Create the Stacked bar chart

```python

# Define colors for the blockchains
colors = {
    'Solana': '#A080FF',
    'Tron': '#909090',
    'Ethereum': '#69D2E7',
    'Base': '#8B4513',
    'Arbitrum': '#FFA07A',
    'BNB Chain': '#FFD700',
    'Avalanche': '#E9967A',
    'Optimism': '#DC143C',
    'Polygon POS': '#98FB98',
    'TON': '#B0E0E6',
    'Celo': '#FFFF00'
}

# Order of blockchains for stacking
order = ['Solana', 'Tron', 'Ethereum', 'Base', 'Arbitrum', 'BNB Chain', 'Avalanche', 'Optimism', 'Polygon POS', 'TON', 'Celo']

# --- Matplotlib Code ---

fig, ax = plt.subplots(figsize=(15, 10))
fig.patch.set_facecolor('#f0f0f0')

# Stacked bar chart
bottom = np.zeros(len(df))
for i, blockchain in enumerate(order):
    ax.bar(df.index, df[blockchain], bottom=bottom, color=colors[blockchain], label=blockchain, width=20)
    bottom += df[blockchain]


# Set y-axis formatting
ax.set_ylabel('$ Trillions', fontsize=12)
ax.tick_params(axis='y', labelsize=10)
ax.set_yticks(np.arange(0, 2.5, 0.5))
ax.set_yticklabels([f'{i}' for i in np.arange(0, 2.5, 0.5)], fontsize=10)
ax.set_ylim(0, 3)

# Set x-axis formatting
ax.set_xlabel('', fontsize=12)
ax.tick_params(axis='x', labelsize=10)
plt.xticks(df.index, [date.strftime('%b-%y') for date in df.index], rotation=45, ha='right')

# Title
ax.set_title('Stablecoin Volume By Blockchain', fontsize=14)

# Legend
ax.legend(loc='upper left', fontsize=10, ncol=6, frameon=False)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_linewidth(0.5)
ax.spines['bottom'].set_color('black')

ax.grid(visible=None)

plt.tight_layout()
plt.show()
```
