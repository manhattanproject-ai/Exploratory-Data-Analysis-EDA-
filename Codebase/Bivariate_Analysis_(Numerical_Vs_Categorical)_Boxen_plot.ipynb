# 1. Import necessary dependencies

```python
# libraries & dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_theme(style="darkgrid")
```

# 2. Create the dataset

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.shape
```

# 3. Boxen plot visualization for numerical vs categorical features

### A. Boxen plot for species vs sepal_length

```python
# Create the boxen plot for sepal_length by species
plt.figure(figsize=(8, 6))
sns.boxenplot(x='species', y='sepal_length', data=iris, palette="viridis")
plt.title("sepal_length by species")
plt.xlabel("species")
plt.ylabel("sepal_length")
plt.show()
```

### B. Boxen plot for all numerical features with categorical variable

```python
# --- Boxen Plots for Numerical vs Categorical Variables ---

numerical_features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
categorical_feature = 'species'

num_plots = len(numerical_features)
rows = (num_plots + 2) // 3
cols = min(num_plots, 2)

fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
fig.suptitle("Boxen Plots of Iris Features by Species", fontsize=16, y=1.02)

axes = axes.flatten()

for i, numerical_col in enumerate(numerical_features):
    if i < len(axes):
        ax = axes[i]
        sns.boxenplot(x=categorical_feature, y=numerical_col, data=iris, ax=ax, palette="viridis")
        ax.set_title(f"{numerical_col} by {categorical_feature}")
        ax.set_xlabel(categorical_feature)
        ax.set_ylabel(numerical_col)

# Remove any unused subplots
if num_plots < len(axes):
    for j in range(num_plots, len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Interpretation of Boxen plot

Interpreting the Attached Boxen Plots:

1. Sepal Length by Species:

* Setosa: Shows a relatively narrow distribution with most data concentrated in the central boxes. The outer boxes are also quite compact, indicating less extreme values.
* Versicolor: Exhibits a wider distribution than setosa, with the boxes extending over a larger range. The outer boxes are also more spread out, suggesting more variability in the tails.
* Virginica: Has the widest distribution, with the boxes covering the largest range of sepal lengths. The outer boxes are also quite extended, indicating more extreme high and low values compared to the other species.
* Comparison: Setosa has the shortest and least variable sepal lengths. Virginica has the longest and most variable sepal lengths, with versicolor falling in between. The boxen plots highlight the increasing spread in sepal length as we move from setosa to virginica.

2. Sepal Width by Species:

* Setosa: Shows a distribution centered at a higher sepal width compared to the others. The boxes are relatively wide, indicating a good amount of data in the central quantiles. The outer boxes are somewhat spread.
* Versicolor: Has a distribution centered at a lower sepal width than setosa, with the boxes spanning a narrower range than setosa. The outer boxes are also less extended.
* Virginica: Shows a distribution with a median similar to versicolor but with slightly wider boxes, indicating more spread in the central quantiles. The outer boxes are also more extended than versicolor's.
* Comparison: Setosa tends to have wider sepals. Versicolor has the narrowest central distribution. Virginica shows more variability in the tails of its sepal width.

3. Petal Length by Species:

* Setosa: Displays a very narrow distribution with highly concentrated data in the central boxes. The outer boxes are extremely compact, indicating very few extreme values.
* Versicolor: Shows a wider distribution than setosa, with the boxes extending over a larger range of petal lengths. The outer boxes are also more spread.
* Virginica: Has the widest distribution, with the boxes covering the largest range of petal lengths. The outer boxes are also the most extended, indicating more variability in the tails.
* Comparison: Petal length is a key differentiator. Setosa has very short and consistent petal lengths. Versicolor and virginica have progressively longer and more variable petal lengths, with the boxen plots clearly showing the increasing spread.

4. Petal Width by Species:

* Setosa: Shows a very narrow distribution with highly concentrated data. The outer boxes are very compact.
* Versicolor: Exhibits a wider distribution than setosa, with more spread in the central boxes and slightly extended outer boxes.
* Virginica: Has the widest distribution, with the boxes covering the largest range of petal widths and the most extended outer boxes.
* Comparison: Similar to petal length, petal width effectively distinguishes the species. Setosa has very narrow and consistent petals. Versicolor and virginica have progressively wider and more variable petal widths, clearly visualized by the increasing spread in the boxen plots.

In Summary:

These boxen plots provide a more detailed view of the distribution of each Iris feature, especially in the tails, compared to standard box plots. They reinforce the findings from box plots but give a better sense of the shape and spread across different quantiles:

* Petal length and petal width are the most distinct features, with setosa showing very narrow distributions and clear separation from the wider distributions of versicolor and virginica.
* Sepal length also shows separation, with increasing length and variability from setosa to virginica.
* Sepal width has the most overlap, with setosa tending to have wider sepals, but the distributions are less clearly separated.

# 4. Create new dataset

```python
wine_data_red= "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
wine_df = pd.read_csv(wine_data_red, sep=';')
```

# 5. Boxen plot visualization for numerical vs categorical features

### A. Boxen plot for quality vs citric acid

```python
# Create the Boxen plot for citric acid by quality
plt.figure(figsize=(15, 6))
sns.boxenplot(x='quality', y='citric acid', data=wine_df, palette="viridis")
plt.title("citric acid by quality")
plt.xlabel("quality")
plt.ylabel("citric acid")
plt.show()
```

### B. Boxen plot for all numerical features with categorical variable

```python
# --- Boxen Plots for Numerical vs Categorical Variables ---

numerical_features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']
categorical_feature = 'quality'

num_plots = len(numerical_features)
rows = (num_plots + 2) // 3
cols = min(num_plots, 4)

fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
fig.suptitle("Boxen Plots of Wine Features by quality", fontsize=16, y=1.02)

axes = axes.flatten()

for i, numerical_col in enumerate(numerical_features):
    if i < len(axes):
        ax = axes[i]
        sns.boxenplot(x=categorical_feature, y=numerical_col, data=wine_df, ax=ax, palette="viridis")
        ax.set_title(f"{numerical_col} by {categorical_feature}")
        ax.set_xlabel(categorical_feature)
        ax.set_ylabel(numerical_col)

# Remove any unused subplots
if num_plots < len(axes):
    for j in range(num_plots, len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Interpreting Individual Boxen Plots:

* Fixed Acidity by Quality: The median fixed acidity appears relatively stable across quality ratings, with perhaps a slight increase at higher qualities. The spread is also fairly consistent.
* Volatile Acidity by Quality: There's a clear negative trend. As wine quality increases, the median volatile acidity significantly decreases. The spread also narrows for higher quality wines, indicating less variability at lower volatile acidity levels.
* Citric Acid by Quality: The median citric acid generally shows a positive trend with increasing quality, although it might plateau at the highest quality. Higher quality wines tend to have a distribution shifted towards higher citric acid values.
* Residual Sugar by Quality: The relationship is complex and doesn't show a strong linear trend in the median. Lower quality wines seem to have a wider range of residual sugar. The distributions are often skewed.
* Chlorides by Quality: A negative trend is evident. Higher quality wines tend to have lower median chloride levels. The spread also appears to decrease for higher quality wines.
* Free Sulfur Dioxide by Quality: The median free sulfur dioxide is relatively stable across quality ratings. The spread seems somewhat consistent, though there might be slightly less variability in the very highest quality.
* Total Sulfur Dioxide by Quality: There's a slight negative trend in the median total sulfur dioxide as quality increases, but it's not very strong. The distributions are generally right-skewed, with some high values across all qualities.
* Density by Quality: The median density is very consistent across quality ratings, with a minimal decrease at the highest quality. The spread is also quite narrow for all quality levels.
* pH by Quality: The median pH shows a slight positive trend with increasing quality, suggesting higher quality wines tend to be slightly less acidic (higher pH). The spread is relatively consistent.
* Sulphates by Quality: A positive trend is noticeable. Higher quality wines tend to have higher median sulphate levels. The distributions shift towards higher sulphate values as quality increases.
* Alcohol by Quality: There's a clear positive correlation. As quality increases, the median alcohol content is generally higher. The distributions are centered at progressively higher alcohol percentages.

Key Takeaways:

* Stronger Indicators of Higher Quality: Lower volatile acidity, lower chlorides, higher citric acid, higher sulphates, and higher alcohol content.
Weaker or More Complex Relationships: Fixed acidity, residual sugar, free sulfur dioxide, total sulfur dioxide, and pH.
* Detailed Spread Information: The boxen plots provide a richer understanding of how the distribution of each feature changes across quality levels, particularly in the tails, compared to standard box plots. This allows for a more nuanced assessment of the variability and potential outliers at each quality rating.
