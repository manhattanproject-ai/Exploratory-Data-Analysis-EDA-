# 3. Create clustered bar chart overtime

```python
# --- Matplotlib Code ---

fig, ax = plt.subplots(figsize=(20, 10))

cmap = plt.cm.viridis  # You can choose a different colormap
im = ax.imshow(df.values, cmap=cmap, aspect='auto', vmin=0, vmax=1)

# Set labels
ax.set_xticks(np.arange(len(technologies)))
ax.set_yticks(np.arange(len(technologies)))
ax.set_xticklabels(technologies, rotation=90, ha="right", fontsize=9)
ax.set_yticklabels(technologies, fontsize=9)

# Title and axis labels
ax.set_title("The Technology Convergence matrix illustrates the relationships between and among technologies.", fontsize=12)
ax.set_xlabel("Catalyzing Technology", fontsize=15)
ax.set_ylabel("Technology", fontsize=15)

# Create colorbar
cbar = fig.colorbar(im, ax=ax, shrink=0.8, location='right', label='Convergence Score')
cbar.ax.tick_params(labelsize=8)

# --- Create custom legend based on the image ---
convergence_levels = [0.8, 0.6, 0.4, 0.2, 0.05] # Approximate values
convergence_labels = ['Highest:', 'High:', 'Mid:', 'Low:', 'Lowest:']
custom_cmap_colors = [cmap(1.0), cmap(0.7), cmap(0.5), cmap(0.3), cmap(0.1)] # Corresponding colors

handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in custom_cmap_colors]
plt.legend(handles, convergence_labels, loc='upper right', bbox_to_anchor=(1.3, 1), frameon=False, fontsize=8)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.grid(visible=None)

# Improve layout
plt.tight_layout(rect=[0, 0, 0.85, 1]) # Adjust layout to make space for the legend
plt.show()
```
